# BOXPLOT 

```python
plt.figure(figsize=(10,8))
sns.boxplot(x= "Age", y= "Industry", data=df, showmeans= True)
plt.show()
```

```python
plt.figure(figsize=(30,10))
sns.swarmplot(x= "Industry", y= "Age", data=df)
plt.show()
```

```python
plt.figure(figsize=(30,10))
sns.violinplot(x= "Industry", y= "Age", data=df)
plt.show()
```

# SCATTERPLOT`

```python
plt.figure(figsize=(30,10))
sns.pairplot(data=df,diag_kws={'bins':6})
plt.show()
```
